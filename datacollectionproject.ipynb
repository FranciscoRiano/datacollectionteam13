```python
u = "https://www.mbastudies.com/EMBA/USA/"
```

```python
response_2 = requests.get(u)
response_2
```

```python
html_1 = response_2.content
html_1
```

```python
USING THE BEAUTIFULSOUP LIBRARY
```

```python
soup_1 = BeautifulSoup(html_1, 'lxml') 
```

EXTRACTING THE NAME OF EACH ACADEMIC EMBA PROGRAM

```python
EMBA_USA_Program = soup_1.find_all('div', {'class': 'program_title'})
EMBA_USA_Program
```

```python
Program_name = [div.find("h4")for div in EMBA_USA_Program]
Program_name
```

```python
Program_name_1 = [name.find("a").string for name in Program_name]
Program_name_1
```

EXTRACTING THE NAME OF EACH UNIVERSITY OR SCHOOL

```python
EMBA_USA_School = soup_1.find_all('div', {'class': 'school'})
EMBA_USA_School
```

```python
Name_university = [div.string for div in EMBA_USA_School]
Name_university
```
